```python
from pathlib import Path
from product_bundle.dataset import read_dataset
from product_bundle.utils import load_product_df, load_all_bundles
import pickle

data_dir = Path("../data/")
df = read_dataset(data_dir)
product_df = load_product_df(data_dir)

with open(f"../static/content_bundles.pickle", "rb") as pickle_file:
    content_bundles = pickle.load(pickle_file)

with open(f"../static/rule_bundles.pickle", "rb") as pickle_file:
    rule_bundles = pickle.load(pickle_file)

with open(f"../static/collaborative_bundles.pickle", "rb") as pickle_file:
    collaborative_bundles = pickle.load(pickle_file)

actual_bundles = set(df.groupby(["InvoiceNo"])["StockCode"].apply(frozenset).to_list())
```
